Я сохранил 5 страниц с сайта кинопоиска с выдачей по 200 фильмов и распарсил их.

```python
import requests
from bs4 import BeautifulSoup

#тут будут храниться названия фильмов
films = []

def read_file(filename):
    with open(filename) as input_file:
        text = input_file.read()
    return text

def parse_kinopoisk(filename):
    
    text = read_file(filename)
    soup = BeautifulSoup(text)
    film_list = soup.find('div', {'class': 'tenItems'})
    items = film_list.find_all('div', {'class': 'item _NO_HIGHLIGHT_'})
    for item in items:
        name = item.find('div', {'class': 'name'}).find('a').text
        films.append(name)
    return films   

parse_kinopoisk('1.html')
parse_kinopoisk('2.html')
parse_kinopoisk('3.html')
parse_kinopoisk('4.html')
parse_kinopoisk('5.html')
```

```python
Как проверить, есть ли фильм на платформе или нет? Я вводил поисковый запрос на платформе и если страница выдает, 
что фильм не найден, то я записываю 0, иначе 1. Понятно, что подход грубоват, так как платформы ivi и megago устроены
так, что выдают фильм даже при сопадении по одному слову. Его можно улучшить, если среди всей выдачи(если она есть), 
проверять каждый фильм с нашим на совпадение по названию. А вот сам поиск платформы улучшить мы не можем, если вдруг
по исходному навзаниею он выдает пустоту, из-за того что название на кинопоиске длиннее, например.
```

```python
Теперь проверим есть ли фильмы на сайте ivi таким же образом, как искали навзания фильмов,
тут капча не вылезает и можно не сохранять файлы вручную.
```

```python
f = films
```

Сначала почистим названия удалив 'ТВ', 'видео'.

```python
spis = []
for x in f:
    if x.find('(ТВ)')!=-1:
        x = x.replace('(ТВ)','')
    if x.find('(видео)')!=-1:
        x = x.replace('(видео)','')
    spis.append(x)    
```

```python
print(spis)
```

```python
y = []
def parse_ivi(filename):
    text = read_file(filename)
    soup = BeautifulSoup(text)
    
    poisk = soup.find('div', {'class': 'content-main'})
    if poisk.find('h2')==None:
        y.append(1)
    else:
        y.append(0)
             
    return y
    

for i in range (0,len(films)):
    url = 'https://www.ivi.ru/search/?q= ' + spis[i]
    r = requests.get(url, headers = headers)
    with open('test.html', 'wb') as output_file:
        output_file.write(r.text.encode('utf-8'))
    parse_ivi('test.html')    
 
```

```python
Теперь проверим есть ли фильмы на сайте megogo.ru, тут капча не вылезает и тоже все ок.
```

```python
z = []
def parse_megogo(filename):
    text = read_file(filename)
    soup = BeautifulSoup(text)
    
    poisk = soup.find('div', {'class': "not-found"})
    if poisk==None:
        z.append(1)
    else:
        z.append(0)
             
    return z
    

for i in range (0,len(films)):
    url = 'https://megogo.ru/ru/search-extended?q=' + spis[i]
    r = requests.get(url, headers = headers)
    with open('test.html', 'wb') as output_file:
        output_file.write(r.text.encode('utf-8'))
    parse_megogo('test.html')     
```

Теперь составим таблицу, 1 если фильм есть на данной платформе, 0 - нет.

```python
import pandas as pd
df = pd.DataFrame({ 'Фильмы':  spis, 'ivi.ru': y, 'megogo.ru': z})
df.head()
```

```python
#запись таблицы в csv файл
df.to_csv('table.csv',encoding = 'utf-8', sep=',', index=False)
```
